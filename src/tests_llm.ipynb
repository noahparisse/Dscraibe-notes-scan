## Transformers

### Pipeline

```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
classifier(["J'adore ce cours", "Je te déteste"])
```

```python
classifier = pipeline("zero-shot-classification")
classifier("J'adore ce bon petit repas bien épicé", candidate_labels=["nourriture, education, sport"])
```

```python
classifier = pipeline("text-generation")
classifier("Ce soir je veux")
```
